```python
def instruction(s: str) -> int:
    """This instruction takes a string s of the form (R/L)# and returns a positive/negative number based on R/L.

    >>> instruction("L68")
    -68
    """

    if s[0] == "L":
        return -int(s[1:])
    elif s[0] == "R":
        return int(s[1:])
    return f"Invalid instruction {s}."
```

```python
def codebreaking(instructions: List[str]) -> int:
    """This function will receive as input a List of instructions in the form L/R followed by an integer number and output the number of times the circular safe lock stops at 0. The lock is modular to 100.

    === Preconditions ===
    - The dial starts pointing at 50.
    - The List of instructions contains string s formatted as s[0] = R/L and s[1:] is some number.

    >>> instructions = ["L68", "L30", "R48", "L5", "R60", "L55", "L1", "L99", "R14", "L82"]
    >>> codebreaking(instructions)
    3
    """
    counter = 0
    dial_position = 50
    for i in instructions:
        dial_position = (dial_position + instruction(i)) % 100
        if dial_position == 0:
            counter += 1

    return f"The number of times the dial turned to 0 is {counter}"
```

```python
"""We must open the file to retrieve all the instructions so I am building a function to do this."""

def load_instructions(filename: str) -> List[int]:
    with open(filename, "r") as f:
        instructions = []
        for line in f.readlines():
            instructions.append(line.strip())

    return instructions
```

```python
"""Here we will break the code using the instructions below and put all the functions together."""
instructions = load_instructions("./day1puzzle1input.txt")
codebreaking(instructions)
```
